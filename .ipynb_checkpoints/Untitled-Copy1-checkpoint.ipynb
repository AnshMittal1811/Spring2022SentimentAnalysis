```python
def dataset_extraction(data):
    df = []
    crypto = ['shib', 'doge', 'shiba', 'dogecoin']
    
    for i in tqdm(range(len(data['messages']))):
        frame = []
        st = data['messages'][i]
        
        if type(st['text']) == str:
            if any(curr in st['text'].lower() for curr in crypto):
                frame.append(st['date'].split("T")[0])
                frame.append(st['text'])

        # Start from here.
        # Let's say we have a list here: [" dsbhfvmjd", "dfbvjuhdm", 
        # {type: "dftgvudk", text: "gvhvsdmhc"}]
        if type(st['text']) == list: 
            b = str()
            
            for j in st['text']: 
                if type(j) == dict:
                    a = j['text']
                    b += a
                else: 
                    b += j
                    
            if any(curr in b.lower() for curr in crypto):
                frame.append(st['date'].split("T")[0])
                frame.append(b)
        
        if len(frame): 
            df.append(frame)
        

    return df
```

```python
messages = dataset_extraction(data)
```

```python
# def detect_language(messages):
#     df_en, non_df_en = [], []
    
#     for mess in messages:
#         print(mess[1])
        
#         detector = Translator()
#         dec = detector.detect(mess[1])
#         if dec.lang =='en':
#             df_en.append(mess)
#         else: 
#             non_df_en.append(mess)
#     print(df_en, non_df_en)
#     return df_en, non_df_en

def preprocessing(df): 
    df = remove_case_sensitive(df)
    df = remove_htmls_and_urls(df)
    df = remove_extra_spaces_between_words(df)
    df = perform_contractions(df)
    df = remove_stop_words(df)
    df = lemmatization(df)
    return df
    
    
def remove_case_sensitive(dataframe):
    dataframe['Messages'] = dataframe['Messages'].apply(lambda txt: str(txt).lower())
    return dataframe

def remove_htmls_and_urls(dataframe):
    dataframe['Messages'] = dataframe['Messages'].apply(lambda txt: re.sub(r"http\S+", "", txt) )
    dataframe['Messages'] = dataframe['Messages'].apply(lambda txt: BeautifulSoup(txt, 'lxml').get_text())
    return dataframe

def remove_extra_spaces_between_words(dataframe): 
    dataframe['Messages'] = dataframe['Messages'].apply(lambda txt: re.sub(" +"," ", txt))
    return dataframe

def perform_contractions(dataframe):
    dataframe['Messages'] = dataframe['Messages'].apply(lambda txt: contractions.fix(txt))
    return dataframe

def remove_stop_words(dataframe): 
    def Remove_Stopwords(message):
        tokens = message.split(" ")
        stop_words_list = set (stopwords.words('english'))
        clean_message = [word for word in tokens if not word in stop_words_list]
        return [(" ").join(clean_message)]

    dataframe['Messages'] = dataframe['Messages'].apply(lambda txt: Remove_Stopwords(txt))
    return dataframe

def lemmatization(dataframe): 
    def lemmatize(message):
#         print(message)
        word_lemma = WordNetLemmatizer()
        lemmatize_word = [word_lemma.lemmatize(word) for word in message]
        return (" ").join(lemmatize_word)
    
    dataframe['Messages']= dataframe['Messages'].apply(lambda txt: lemmatize(txt))
    return dataframe
```

```python
def wordnet_based_differentiation(dataframe):
    def word_in_english(message): 
        
        st = []
        for word in message.split(" "): 
            if wordnet.synsets(word) or word.lower() in ["doge", "dogecoin", "shib", "shiba"]:
                st.append(1)
            else: 
                st.append(0)
            
        if sum(st)/len(st) >= 0.5: 
            return message
        else: 
            return " "
                    
    dataframe['Messages'] = dataframe['Messages'].apply(lambda txt: word_in_english(txt))
                
    return dataframe
```
